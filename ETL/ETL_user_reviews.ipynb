```python
# Se importan las librerías necesarias

import pandas as pd
import numpy as np
import ast
import seaborn as sns
import matplotlib.pyplot as plt
import json
from pandas import json_normalize
import nltk
```

```python
# Se crea una funcion para cargar archivos JSON
def load_json_lines(file_path):

    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(ast.literal_eval(line))
    return pd.DataFrame(data)
```

```python
#Funcion para eliminar reviews duplicadas del DataFrame
def eliminar_duplicados(df_reviews):

    #Normaliza los valores de las columnas "reviews"
    normalized_reviews = json_normalize(df_reviews['reviews'])

    # Añade la columna "user_id" al DataFrame "normalized_reviews"
    normalized_reviews = pd.concat([df_reviews['user_id'], normalized_reviews], axis=1)

    # Elimina los duplicados del DataFrame "normalized_reviews" basado en "user_id"
    user_reviews = normalized_reviews[~normalized_reviews.duplicated(subset='user_id')]

    # Devuelve el DataFrame creado
    return user_reviews

# Se utiliza la función y devuelve el resultado en un nuevo DataFrame
user_reviews = eliminar_duplicados(df_reviews)

# Se muestra el DataFrame
user_reviews
```

# *Feature Engineering*

Se incluirá la columna "sentiment_analysis" la cual aplicará análisis de sentimiento con NLP.Este toma el valor "0" si la reseña es negativa, "1" si la reseña es neutral y "2" si la reseña es positiva.De no poder realizarse este análisis, tomará el valor "1". Esta nueva columna reemplazará a la de "review" para facilitar el trabajo de los modelos de ML y análisis de datos.



```python
# Transformo todas las letras de la columna "review" a minuscula
df_user_reviews['review'] = df_user_reviews['review'].str.lower()
```

```python
# Importo la libreria "re"
import re

# Se buscan y eliminan los caracteres especiales haciendo uso de la libreria re
df_user_reviews['review'] = df_user_reviews['review'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', str(x)) if pd.notnull(x) else '')
```

```python
from nltk.corpus import stopwords

# Se descarga la lista de stopwords
nltk.download('stopwords')

# Crea un conjunto que contiene las stopwords en inglés
stop_words = set(stopwords.words('english'))

# Elimino las stopwords
df_user_reviews['review'] = df_user_reviews['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]) if pd.notnull(x) else '')
```

```python
from nltk.stem import WordNetLemmatizer, PorterStemmer

# Se descargan datos necesarios para lematización desde WordNet
nltk.download('wordnet')

# Crea un objeto WordNetLemmatizer para lematizar palabras
lemmatizer = WordNetLemmatizer()

# Lematizo cada elemento de la columna "review"
df_user_reviews['review'] = df_user_reviews['review'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]) if pd.notnull(x) else '')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Descarga el léxico necesario
nltk.download('vader_lexicon')

# Crea un objeto SentimentIntensityAnalyzer de NLTK que se utilizará para realizar el análisis de sentimientos
sia = SentimentIntensityAnalyzer()

# Aplica el analisis de sentimiento a cada elemento de "review"
df_user_reviews['sentiment_analysis'] = df_user_reviews['review'].apply(lambda x: sia.polarity_scores(x)['compound'] if pd.notnull(x) else None)

# Asignar etiquetas de sentimiento (0 para negativo, 1 para neutral, 2 para positivo)
df_user_reviews['sentiment_label'] = pd.cut(df_user_reviews['sentiment_analysis'], bins=[-1, -0.1, 0.1, 1], labels=[0, 1, 2])
```

```python
# Elimino la columna "review" ya que queda sin uso
df_user_reviews = df_user_reviews.drop('review', axis=1)

# Muestro el DF final
df_user_reviews
```

```python
# Exporto el DataFrame en formato CSV, JSON y Parquet

df_user_reviews.to_csv('df_user_reviews.csv', index=False)

df_user_reviews.to_json('df_user_reviews.json', orient='records', lines=True)

df_user_reviews.to_parquet('df_user_reviews.parquet', index=False)
```
