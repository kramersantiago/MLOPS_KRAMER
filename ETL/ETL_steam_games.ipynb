```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from pandas import json_normalize
import seaborn as sns
import ast
```

```python
# Se crea una funcion para cargar archivos JSON
def load_json_lines(file_path):
    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            try:
                json_object = json.loads(line)
                data.append(json_object)
            except json.JSONDecodeError as e:
                print(f"Error decoding JSON on line: {line}. Error: {e}")

    return pd.DataFrame(data)
```

```python
# Cargar el archivo JSON
df_steam_games = load_json_lines(r'C:\Users\Renato\Desktop\MLOPS_KRAMER\PI MLOps - STEAM\output_steam_games.json')

# Mostrar el DF
df_steam_games
```

```python
print(df_steam_games.info)

df_steam_games.describe()
```

```python
# Chequeo de nulos por columna
print(df_steam_games.isna().sum())
```

```python
# Eliminar valores nulos
df_steam_games = df_steam_games.dropna(how='all')

# Mostrar el DF
df_steam_games
```

```python
# Valores unicos de la columna "price"
df_steam_games['price'].unique()
```

```python
# Creo una mascara para filtrar los valores que contengan "Free"
mask_price = df_steam_games['price'].astype(str).str.contains('Free', case=False)

# Reemplazo los valores que cumplan la condicion por 0
df_steam_games['price'] = df_steam_games['price'].where(~mask_price, 0)

# Muestro el DF
df_steam_games
```

```python
# Elimina filas con valores nulos en la columna "price"
df_steam_games = df_steam_games[df_steam_games['price'].notnull()]

# Elimina filas con valores no numericos en la columna "price"
df_steam_games = df_steam_games[pd.to_numeric(df_steam_games['price'], errors='coerce').notnull()]

# Muestra los valores unicos para asegurarse que no hayan valores no numericos
print(df_steam_games['price'].unique())

# Muestra el tamaño del DataFrame
print('\n', df_steam_games.shape)

# Muestra el DF
df_steam_games.head()
```

```python
# Muestra el porcentaje de datos nulos por columna
porcentaje_nulos = (df_steam_games.isnull().mean() * 100).round(2)

porcentaje_nulos
```

```python
# Comprobar si "developer" está vacía y copiar desde "publisher"
mask_developer_vacio = df_steam_games['developer'].isnull()
df_steam_games.loc[mask_developer_vacio, 'developer'] = df_steam_games.loc[mask_developer_vacio, 'publisher']

# Comprobar si "publisher" está vacía y copiar desde "developer"
mask_publisher_vacio = df_steam_games['publisher'].isnull()
df_steam_games.loc[mask_publisher_vacio, 'publisher'] = df_steam_games.loc[mask_publisher_vacio, 'developer']

# Chequeo de nulos
(df_steam_games.isnull().mean() * 100).round(2)
```

```python
# Se convierte la columna "release_date" a formato fecha
df_steam_games['release_date'] = pd.to_datetime(df_steam_games['release_date'], errors='coerce', exact=False)

df_steam_games
```

```python
# Exporto los DataFrame en formato CSV, JSON y Parquet

df_steam_games.to_csv('df_steam_games.csv', index=False)

df_steam_games.to_json('df_steam_games.json', orient='records', lines=True)

df_steam_games.to_parquet('df_steam_games.parquet', index=False)
```
