```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
from pandas import json_normalize
import seaborn as sns
import ast
```

```python
# Se crea una funcion para cargar archivos JSON
def load_json_lines(file_path):

    data = []
    with open(file_path, 'r', encoding='utf-8') as file:
        for line in file:
            data.append(ast.literal_eval(line))
    return pd.DataFrame(data)
```

```python
# Cargar el archivo JSON
df_items = load_json_lines(r'C:\Users\Renato\Desktop\MLOPS_KRAMER\PI MLOps - STEAM\australian_users_items.json')

# Mostrar el DF
df_items
```

```python
# Da una informacion sobre el DataFrame creado
print(df_items.info)

# Información estadística de las columnas en el DataFrame
df_items.describe()
```

```python
# Verificación de datos nulos
print((df_items.isnull().mean() * 100).round(2))
```

```python
# Lista que almacena los DataFrames normalizados
lista_df = []

# Itera sobre las filas de la columna 'items'
for index, fila in df_items['items'].items():

    # Normaliza la fila actual y agrega el DataFrame resultante a la lista
    df_normalizado = pd.json_normalize(fila)
    lista_df.append(df_normalizado)
```

```python
# Concatena los DataFrames normalizados en uno
data_items = pd.concat(lista_df, ignore_index=True)

# Muestra el DF
data_items
```

```python
# Elimina la columna items anidada del DataFrame
df_items.drop(columns=['items'], inplace=True)

# Concatena el DataFrame original con la versión aplanada de la columna items
df_items_normalizado = pd.concat([df_items, data_items], axis=1)

# Muestra el DataFrame
df_items_normalizado
```

```python
# Chequeo de valores NaN por columna
cantidad_NaN = df_items_normalizado.isna().sum()

print(cantidad_NaN)

# Porcentaje de datos nulos por columna
porcentaje_nulos = (df_items_normalizado.isnull().mean() * 100).round(2)

print('\n', porcentaje_nulos)

df_user_items = df_items_normalizado
```

```python
# Eliminar los datos nulos
df_user_items = df_user_items.dropna()

df_user_items
```

```python
# Exporto los DataFrame en formato CSV, JSON y Parquet

df_user_items.to_csv('df_user_items.csv', index=False)

df_user_items.to_json('df_user_items.json', orient='records', lines=True)

df_user_items.to_parquet('df_user_items.parquet', index=False)
```
