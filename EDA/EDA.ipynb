# Union de DataFrames y EDA mas completo

Se unen los DataFrames "df_user_items" y "df_steam_games" utilizando en comun la columna "item_id" para realizar un EDA mas profundo con graficos para analizar patrones y comportamientos.

```python
df_steam_games = df_steam_games.rename(columns={'id': 'item_id'})

df_combined = pd.merge(df_user_items, df_steam_games, on='item_id', how='inner')
df_combined.head()
```

```python
# Agrupar por "user_id" y calcular la cantidad total de items para cada usuario
user_items_count = df_combined.groupby('user_id')['items_count'].sum()
custom_bins = [0, 20, 40, 60, 80, 100, 120, 140]

# Visualizar la distribución de la cantidad de items por usuario
plt.figure(figsize=(10, 6))
plt.hist(user_items_count, bins=custom_bins, color='skyblue', edgecolor='black', width=10)
plt.title('Distribución de la cantidad de items por usuario')
plt.xlabel('Cantidad de items')
plt.ylabel('Número de usuarios')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Desglosa las listas en la columna 'genres'
genre_counts = df_steam_games['genres'].explode().value_counts()

# Gráfico de barras para los géneros más populares
plt.figure(figsize=(12, 6))
genre_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Géneros de juegos más populares')
plt.xlabel('Género')
plt.ylabel('Número de juegos')
plt.grid(axis='y', alpha=0.75)
plt.xticks(rotation=45, ha='right')
plt.show()
```

# EDA df_user_reviews

```python
# Conteo de reseñas
review_counts = df_user_reviews['recommend'].value_counts()
print(review_counts)
# Gráfico de barras para la distribución de reseñas
plt.figure(figsize=(8, 5))
review_counts.plot(kind='bar', color=['green', 'red'], alpha=0.8)
plt.title('Distribución de reviews')
plt.xlabel('Recomendación')
plt.ylabel('Número de reviews')
plt.xticks([0, 1], ['Recomendado', 'No recomendado'], rotation=0)
plt.show()
```

```python
# Matriz de correlación entre "recommend" y "sentiment_label"
correlation_matrix = df_user_reviews[['recommend', 'sentiment_label']].corr()

# Mapa de calor para visualizar la correlación
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlación entre recomendación y sentimiento')
plt.show()
```

```python
# Visualizar la distribución de 'playtime_forever'
plt.figure(figsize=(12, 6))
sns.histplot(df_user_items['playtime_forever'], bins=50, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribución de Playtime Forever')
plt.xlabel('Playtime Forever')
plt.ylabel('Número de Usuarios')
plt.grid(axis='y', alpha=0.75)
plt.show()

# Visualizar la distribución de 'playtime_2weeks'
plt.figure(figsize=(12, 6))
sns.histplot(df_user_items['playtime_2weeks'], bins=50, kde=True, color='orange', edgecolor='black')
plt.title('Distribución de Playtime en las Últimas 2 Semanas')
plt.xlabel('Playtime en las Últimas 2 Semanas')
plt.ylabel('Número de Usuarios')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
